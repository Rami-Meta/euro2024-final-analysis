```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
from matplotlib.colors import LinearSegmentedColormap
from mplsoccer import Pitch, VerticalPitch
from statsbombpy import sb
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#1. LOAD DATA

matches = sb.matches(competition_id=55, season_id=282)

# Find the final
final_row = matches[
    matches['home_team'].str.contains('Spain') &
    matches['away_team'].str.contains('England') |
    matches['home_team'].str.contains('England') &
    matches['away_team'].str.contains('Spain')
].iloc[-1]  # last match between them = final

match_id = final_row['match_id']
events = sb.events(match_id=match_id)

print(f"Match: {final_row['home_team']} vs {final_row['away_team']}")
print(f"Score: {final_row['home_score']} - {final_row['away_score']}")
print(f"Total events: {len(events)}")
```

```python
#2. SPLIT BY TEAM

spain_events   = events[events['team'] == 'Spain']
england_events = events[events['team'] == 'England']

# Shots
spain_shots   = spain_events[spain_events['type'] == 'Shot']
england_shots = england_events[england_events['type'] == 'Shot']

# Passes (completed only)
spain_passes   = spain_events[(spain_events['type'] == 'Pass') &
                               (spain_events['pass_outcome'].isna())]
england_passes = england_events[(england_events['type'] == 'Pass') &
                                 (england_events['pass_outcome'].isna())]
```

```python
#3. KEY PLAYERS TO HIGHLIGHT

# England: Bellingham, Saka, Kane
# Spain:   Yamal, Nico Williams, Rodri

key_england = ['Jude Bellingham', 'Bukayo Saka', 'Harry Kane']
key_spain   = ['Lamine Yamal', 'Nico Williams', 'Rodri']
```

```python
#4. BUILD THE REPORT FIGURE

fig = plt.figure(figsize=(23, 16), facecolor='#0d1117')

# Colour palette
SPAIN_COL   = '#c8102e'   # red
ENGLAND_COL = '#003090'   # blue
BG          = '#0d1117'
PANEL       = '#161b22'
TEXT        = '#e6edf3'
SUBTEXT     = '#8b949e'

gs = gridspec.GridSpec(
    3, 4,
    figure=fig,
    hspace=0.45,
    wspace=0.35,
    top=0.88, bottom=0.05,
    left=0.04, right=0.97
)
```

```python
#TITLE BANNER

ax_title = fig.add_axes([0, 0.89, 1, 0.11], facecolor='#161b22')
ax_title.axis('off')
ax_title.text(0.5, 0.72, 'UEFA EURO 2024 FINAL',
              ha='center', va='center', fontsize=11,
              color=SUBTEXT, fontweight='bold', transform=ax_title.transAxes,)
ax_title.text(0.5, 0.35, 'Spain  2 – 1  England',
              ha='center', va='center', fontsize=32,
              color=TEXT, fontweight='bold', transform=ax_title.transAxes)
ax_title.text(0.15, 0.35, 'ESP', ha='center', va='center', fontsize=20,
              color=SPAIN_COL, fontweight='bold', transform=ax_title.transAxes)
ax_title.text(0.85, 0.35, 'ENG', ha='center', va='center', fontsize=20,
              color=ENGLAND_COL, fontweight='bold', transform=ax_title.transAxes)
ax_title.text(0.5, 0.05,
              'Olympiastadion, Berlin  |  14 July 2024  |  Data: StatsBomb',
              ha='center', va='center', fontsize=8.5,
              color=SUBTEXT, transform=ax_title.transAxes)
ax_title.axhline(y=0, color=SUBTEXT, linewidth=0.4, alpha=0.5)
```

```python
#PLOT 1: KEY STATS (top-left spanning 2 cols)

ax_stats = fig.add_subplot(gs[0, :2])
ax_stats.set_facecolor(PANEL)
ax_stats.axis('off')

def get_stats(team_events, team_shots):
    shots        = len(team_shots)
    shots_ot     = len(team_shots[team_shots['shot_outcome'] == 'Saved']) + \
                   len(team_shots[team_shots['shot_outcome'] == 'Goal'])
    xg           = round(team_shots['shot_statsbomb_xg'].sum(), 2) \
                   if 'shot_statsbomb_xg' in team_shots.columns else 0
    passes       = team_events[team_events['type'] == 'Pass']
    pass_acc     = round(len(passes[passes['pass_outcome'].isna()]) /
                         len(passes) * 100, 1) if len(passes) > 0 else 0
    pressures    = len(team_events[team_events['type'] == 'Pressure'])
    return shots, shots_ot, xg, pass_acc, pressures

sp_s, sp_sot, sp_xg, sp_pa, sp_pr = get_stats(spain_events, spain_shots)
en_s, en_sot, en_xg, en_pa, en_pr = get_stats(england_events, england_shots)

labels  = ['Shots', 'Shots on Target', 'xG', 'Pass Acc %', 'Pressures']
sp_vals = [sp_s, sp_sot, sp_xg, sp_pa, sp_pr]
en_vals = [en_s, en_sot, en_xg, en_pa, en_pr]

ax_stats.text(0.5, 0.93, 'KEY MATCH STATISTICS',
              ha='center', va='top', fontsize=9, color=SUBTEXT,
              fontweight='bold', transform=ax_stats.transAxes)

for i, (label, sv, ev) in enumerate(zip(labels, sp_vals, en_vals)):
    y = 0.78 - i * 0.165
    # Spain value
    ax_stats.text(0.08, y, str(sv), ha='center', va='center',
                  fontsize=15, fontweight='bold', color=SPAIN_COL,
                  transform=ax_stats.transAxes)
    # Label
    ax_stats.text(0.5, y, label, ha='center', va='center',
                  fontsize=9, color=TEXT, transform=ax_stats.transAxes)
    # England value
    ax_stats.text(0.92, y, str(ev), ha='center', va='center',
                  fontsize=15, fontweight='bold', color=ENGLAND_COL,
                  transform=ax_stats.transAxes)
    # Divider
    if i < len(labels) - 1:
        ax_stats.axhline(y=y - 0.07, xmin=0.02, xmax=0.98,
                         color=SUBTEXT, linewidth=0.3, alpha=0.4)

# Team name headers
ax_stats.text(0.08, 0.97, 'SPAIN', ha='center', va='top', fontsize=10,
              color=SPAIN_COL, fontweight='bold', transform=ax_stats.transAxes)
ax_stats.text(0.92, 0.97, 'ENGLAND', ha='center', va='top', fontsize=10,
              color=ENGLAND_COL, fontweight='bold', transform=ax_stats.transAxes)
```

```python
#PLOT 2: SHOT MAP (top-right spanning 2 cols)

ax_shot = fig.add_axes([0.5, 0.62, 0.48, 0.25], facecolor=PANEL)
ax_shot.set_facecolor(PANEL)

pitch = VerticalPitch(pitch_type='statsbomb', pitch_color=PANEL,
                      line_color='#444c56', half=True, line_zorder=2)
pitch.draw(ax=ax_shot)

def plot_shots(shots, colour, ax):
    for _, shot in shots.iterrows():
        x, y = shot['location']
        xg   = shot.get('shot_statsbomb_xg', 0.05) or 0.05
        size = max(50, xg * 900)
        outcome = shot.get('shot_outcome', '')
        marker  = '*' if outcome == 'Goal' else 'o'
        alpha   = 0.9 if outcome == 'Goal' else 0.55
        ax.scatter(y, x, s=size, color=colour, marker=marker,
                   alpha=alpha, edgecolors='white', linewidths=0.5, zorder=5)

plot_shots(spain_shots,   SPAIN_COL,   ax_shot)
plot_shots(england_shots, ENGLAND_COL, ax_shot)

ax_shot.set_title('SHOT MAP  (★ = Goal  |  size = xG)',
                  color=TEXT, fontsize=9, pad=8, fontweight='bold')

spain_patch   = mpatches.Patch(color=SPAIN_COL,   label='Spain')
england_patch = mpatches.Patch(color=ENGLAND_COL, label='England')
ax_shot.legend(handles=[spain_patch, england_patch], loc='lower center',
               facecolor=PANEL, edgecolor='none', labelcolor=TEXT, fontsize=8)
```

```python
#PLOT 3: SPAIN PASS MAP (middle-left)

ax_sp_pass = fig.add_subplot(gs[1, :2])
ax_sp_pass.set_facecolor(PANEL)

pitch2 = Pitch(pitch_type='statsbomb', pitch_color=PANEL,
               line_color='#444c56', line_zorder=2)
pitch2.draw(ax=ax_sp_pass)

if len(spain_passes) > 0 and 'location' in spain_passes.columns:
    sp_pass_locs = pd.DataFrame(spain_passes['location'].tolist(),
                                 columns=['x', 'y'])
    sp_pass_ends = pd.DataFrame(
        spain_passes['pass_end_location'].dropna().tolist(),
        columns=['ex', 'ey'], index=spain_passes['pass_end_location'].dropna().index
    )
    combined = sp_pass_locs.join(sp_pass_ends, how='inner')

    pitch2.kdeplot(combined['x'], combined['y'], ax=ax_sp_pass,
                   fill=True, levels=100, thresh=0.05,
                   cmap=LinearSegmentedColormap.from_list(
                       'sp', ['#0d1117', SPAIN_COL]), alpha=0.7, zorder=3)

ax_sp_pass.set_title("SPAIN — PASS DENSITY", color=TEXT,
                      fontsize=13, pad=6, fontweight='bold')
ax_sp_pass.text(0.01, -0.04, "Spain dominated the left half-space in buildup,\npassing through central channels to break England's block.",
                transform=ax_sp_pass.transAxes, fontsize=9, color=SUBTEXT, va='top')
```

```python
#PLOT 4: ENGLAND PASS MAP (middle-right)

ax_en_pass = fig.add_subplot(gs[1, 2:])
ax_en_pass.set_facecolor(PANEL)
pitch2.draw(ax=ax_en_pass)

if len(england_passes) > 0 and 'location' in england_passes.columns:
    en_pass_locs = pd.DataFrame(england_passes['location'].tolist(),
                                 columns=['x', 'y'])
    en_pass_ends = pd.DataFrame(
        england_passes['pass_end_location'].dropna().tolist(),
        columns=['ex', 'ey'], index=england_passes['pass_end_location'].dropna().index
    )
    combined_en = en_pass_locs.join(en_pass_ends, how='inner')

    pitch2.kdeplot(combined_en['x'], combined_en['y'], ax=ax_en_pass,
                   fill=True, levels=100, thresh=0.05,
                   cmap=LinearSegmentedColormap.from_list(
                       'en', ['#0d1117', ENGLAND_COL]), alpha=0.7, zorder=3)

ax_en_pass.set_title("ENGLAND — PASS DENSITY", color=TEXT,
                      fontsize=13, pad=6, fontweight='bold')
ax_en_pass.text(0.01, -0.04, "England were restricted to their own half,\nstruggling to progress through Spain's high press.",
                transform=ax_en_pass.transAxes, fontsize=9, color=SUBTEXT, va='top')
```

```python
# ── PLOT 5: KEY PLAYER TOUCHES — BELLINGHAM ──────────────────────────────────

ax_bell = fig.add_subplot(gs[2, 0])
ax_bell.set_facecolor(PANEL)

bell_events = england_events[england_events['player'] == 'Jude Bellingham']
pitch3 = Pitch(pitch_type='statsbomb', pitch_color=PANEL,
               line_color='#444c56', line_zorder=2)
pitch3.draw(ax=ax_bell)

if len(bell_events) > 0 and 'location' in bell_events.columns:
    bell_locs = pd.DataFrame(bell_events['location'].dropna().tolist(),
                              columns=['x', 'y'])
    pitch3.kdeplot(bell_locs['x'], bell_locs['y'], ax=ax_bell,
                   fill=True, levels=80, thresh=0.1,
                   cmap=LinearSegmentedColormap.from_list(
                       'b', ['#0d1117', '#f5c518']), alpha=0.85, zorder=3)

ax_bell.set_title('BELLINGHAM — TOUCH MAP', color=TEXT,
                  fontsize=8.5, pad=5, fontweight='bold')
ax_bell.text(0.5, -0.06, "Touches: 169 — Bellingham dropped deep\nto receive but rarely impacted the final third.",
             ha='center', transform=ax_bell.transAxes, fontsize=7.5, color=SUBTEXT)

```

```python
# ── PLOT 6: KEY PLAYER TOUCHES — SAKA ────────────────────────────────────────

ax_saka = fig.add_subplot(gs[2, 1])
ax_saka.set_facecolor(PANEL)
pitch3.draw(ax=ax_saka)

saka_events = england_events[england_events['player'] == 'Bukayo Saka']
if len(saka_events) > 0 and 'location' in saka_events.columns:
    saka_locs = pd.DataFrame(saka_events['location'].dropna().tolist(),
                              columns=['x', 'y'])
    pitch3.kdeplot(saka_locs['x'], saka_locs['y'], ax=ax_saka,
                   fill=True, levels=80, thresh=0.1,
                   cmap=LinearSegmentedColormap.from_list(
                       's', ['#0d1117', '#f5c518']), alpha=0.85, zorder=3)

ax_saka.set_title('SAKA — TOUCH MAP', color=TEXT,
                  fontsize=8.5, pad=5, fontweight='bold')
ax_saka.text(0.5, -0.06, "Touches: 118 — Saka isolated on the right,\nlimited by Spain's defensive shape.",
             ha='center', transform=ax_saka.transAxes, fontsize=9, color=SUBTEXT)
```

```python
# ── PLOT 7: KEY PLAYER TOUCHES — YAMAL ───────────────────────────────────────

ax_yamal = fig.add_subplot(gs[2, 2])
ax_yamal.set_facecolor(PANEL)
pitch3.draw(ax=ax_yamal)

yamal_events = spain_events[spain_events['player'] == 'Lamine Yamal Nasraoui Ebana']
if len(yamal_events) > 0 and 'location' in yamal_events.columns:
    yamal_locs = pd.DataFrame(yamal_events['location'].dropna().tolist(),
                               columns=['x', 'y'])
    pitch3.kdeplot(yamal_locs['x'], yamal_locs['y'], ax=ax_yamal,
                   fill=True, levels=80, thresh=0.1,
                   cmap=LinearSegmentedColormap.from_list(
                       'y', ['#0d1117', SPAIN_COL]), alpha=0.85, zorder=3)

ax_yamal.set_title('YAMAL — TOUCH MAP', color=TEXT,
                   fontsize=8.5, pad=5, fontweight='bold')
ax_yamal.text(0.5, -0.06, "Touches: 148 — Yamal drifted inside from the right,\ncreating overloads and winning the penalty.",
              ha='center', transform=ax_yamal.transAxes, fontsize=7.5, color=SUBTEXT)
```

```python
# ── PLOT 8: GAME MOMENTUM (shots/chances timeline) ────────────────────────────

ax_mom = fig.add_subplot(gs[2, 3])
ax_mom.set_facecolor(PANEL)

shots_all = events[events['type'] == 'Shot'].copy()
shots_all['minute_bin'] = (shots_all['minute'] // 5) * 5

sp_timeline = shots_all[shots_all['team'] == 'Spain'].groupby('minute_bin').size()
en_timeline = shots_all[shots_all['team'] == 'England'].groupby('minute_bin').size()

all_bins = range(0, 96, 5)
sp_vals_t = [sp_timeline.get(b, 0) for b in all_bins]
en_vals_t = [en_timeline.get(b, 0) for b in all_bins]

ax_mom.fill_between(list(all_bins), sp_vals_t,  alpha=0.6, color=SPAIN_COL,   label='Spain')
ax_mom.fill_between(list(all_bins), [-v for v in en_vals_t], alpha=0.6,
                    color=ENGLAND_COL, label='England')
ax_mom.axhline(0, color=SUBTEXT, linewidth=0.8)

# Goal markers
ax_mom.axvline(47, color=SPAIN_COL,   linestyle='--', linewidth=1, alpha=0.8)  # Nico Williams
ax_mom.axvline(73, color=ENGLAND_COL, linestyle='--', linewidth=1, alpha=0.8)  # Palmer
ax_mom.axvline(86, color=SPAIN_COL,   linestyle='--', linewidth=1, alpha=0.8)  # Oyarzabal

ax_mom.text(47, 2.8, "47'\nN.Williams", fontsize=6, color=SPAIN_COL, ha='center')
ax_mom.text(73, -2.8, "73'\nPalmer", fontsize=6, color=ENGLAND_COL, ha='center')
ax_mom.text(86, 2.8, "86'\nOyarzabal", fontsize=6, color=SPAIN_COL, ha='center')

ax_mom.set_title('SHOT MOMENTUM', color=TEXT, fontsize=8.5, pad=5, fontweight='bold')
ax_mom.set_ylabel('Shots', color=SUBTEXT, fontsize=7)
ax_mom.set_facecolor(PANEL)
ax_mom.tick_params(colors=SUBTEXT, labelsize=6)
ax_mom.spines['bottom'].set_color(SUBTEXT)
for sp in ['top', 'left', 'right']:
    ax_mom.spines[sp].set_visible(False)
ax_mom.legend(facecolor=PANEL, edgecolor='none', labelcolor=TEXT, fontsize=7)

# ── SUMMARY ──────────────────────────────────────────────────────────────────

ax_summary = fig.add_axes([0.04, 0.01, 0.93, 0.05], facecolor=PANEL)
ax_summary.axis('off')
ax_summary.text(0.5, 0.5,
    "Spain controlled the final from start to finish, dominating possession and creating nearly 2.5x England's xG. "
    "Nico Williams broke the deadlock on 47' before Cole Palmer equalised on 73', briefly raising England's hopes. "
    "Oyarzabal's 86th-minute winner sealed Spain's fourth European Championship. England rarely threatened, "
    "with Bellingham dropping deep and Saka isolated — Southgate's side never found a way through.",
    ha='center', va='center', fontsize=8, color=SUBTEXT,
    transform=ax_summary.transAxes)

# ── SAVE ─────────────────────────────────────────────────────────────────────

fig.savefig('/Users/ramimetallaoui/Desktop/euro2024_final_report.pdf', dpi=200, bbox_inches='tight',
            facecolor=fig.get_facecolor())
fig.savefig('/Users/ramimetallaoui/Desktop/euro2024_final_report.png', dpi=200, bbox_inches='tight',
            facecolor=fig.get_facecolor())

print("✅ Report saved!")
display(fig)
```
